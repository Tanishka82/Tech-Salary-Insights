## 3. ***Data Wrangling***

###  **üßπData Wrangling Code**

````python
# Salary Cleaning with Hourly/Monthly Filter

# Keep only rows that contain a range with '-'
df = df[df['Salary Estimate'].str.contains('-')]

# Exclude rows with 'Per Hour' or 'Per Year' or other incompatible formats
df = df[~df['Salary Estimate'].str.contains('Per Hour', case=False, na=False)]
df = df[~df['Salary Estimate'].str.contains('Per Year', case=False, na=False)]

# Remove text and symbols
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])  # remove anything in brackets
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Employer Provided Salary:', '', regex=False)
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$', '', regex=False)
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '', regex=False)
df['Salary Estimate'] = df['Salary Estimate'].str.strip()

# Now split into min and max
df[['min_salary', 'max_salary']] = df['Salary Estimate'].str.split('-', expand=True)

# Safely convert to integers
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')

# Drop rows where conversion failed (just in case)
df.dropna(subset=['min_salary', 'max_salary'], inplace=True)

# Compute average salary
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

print("‚úÖ Cleaned salary estimates successfully.")
````

````python
# Handling Outliers & Outlier treatments

# üì¶ Import required library
import numpy as np

# üîç Define a function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"Removing outliers in '{column}' outside [{lower_bound:.2f}, {upper_bound:.2f}]")
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# ‚úÇÔ∏è Treat outliers in avg_salary and Rating
df_cleaned = remove_outliers_iqr(df, 'avg_salary')
df_cleaned = remove_outliers_iqr(df_cleaned, 'Rating')

# üîÅ Replace original df
df = df_cleaned.copy()

# ‚úÖ Check updated shape
print("Updated dataset shape after outlier removal:", df.shape)
````

````python
# üîÅ Rephrase Text
# NOTE: Rephrasing usually requires advanced NLP models like GPT or T5, which are resource-intensive.
# For this example, we‚Äôll use simple synonym replacement using WordNet (as a basic form of rephrasing).

from nltk.corpus import wordnet

def get_synonym(word):
    synonyms = wordnet.synsets(word)
    if synonyms:
        # Return the first synonym‚Äôs lemma if available
        return synonyms[0].lemmas()[0].name()
    return word

def rephrase_text(text):
    words = text.split()
    rephrased_words = [get_synonym(word) for word in words]
    return ' '.join(rephrased_words)

# Apply basic rephrasing
df['Job Description'] = df['Job Description'].apply(rephrase_text)

# Preview the rephrased descriptions
df['Job Description'].head(5)
````

### üîÑ 8. Text Normalization: Stemming & Lemmatization

We apply text normalization techniques to reduce words to their base or root form.

- **Stemming** cuts off suffixes (e.g., ‚Äúrunning‚Äù ‚Üí ‚Äúrun‚Äù), but can be crude.
- **Lemmatization** uses vocabulary and grammar rules to convert words to their dictionary form (e.g., ‚Äúbetter‚Äù ‚Üí ‚Äúgood‚Äù).

For our use case, we prefer **Lemmatization** as it is more accurate and context-aware, especially useful in job descriptions.


````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Define a function to lemmatize tokens
def lemmatize_text(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc]

# Apply lemmatization on the Job Description column
df['Lemmatized Tokens'] = df['Job Description'].astype(str).apply(lemmatize_text)

# Preview
df[['Job Description', 'Lemmatized Tokens']].head(3)
````

#### **üìå Which text normalization technique have you used and why?**

We have used **Lemmatization** for text normalization.

#### **‚úÖ Why Lemmatization?**
- **Context-Aware**: Lemmatization considers the context and converts a word to its meaningful base form (e.g., "running" ‚Üí "run", "better" ‚Üí "good").
- **Grammatically Correct Output**: Unlike stemming, which may return incomplete or non-dictionary words, lemmatization ensures meaningful, clean tokens.
- **Suitable for Professional Text**: Job descriptions are formal and domain-specific. Lemmatization preserves their clarity and semantic meaning, which is important for downstream tasks like classification or keyword extraction.

#### **‚õî Why Not Stemming?**
- **Too Aggressive**: Stemming simply chops off word endings and can result in non-words (e.g., "studies" ‚Üí "studi").
- **Loss of Meaning**: It ignores the context of the word, which may lead to incorrect or unclear results.

Hence, **Lemmatization was chosen** as the more robust and accurate normalization technique for our NLP-based job description analysis.

#### **2. Cross- Validation & Hyperparameter Tuning**

#### **üîÑ ML Model - 3: Support Vector Regressor (SVR) with Hyperparameter Optimization**

To improve our SVR model's performance, we used **GridSearchCV**, a powerful hyperparameter optimization technique that performs an exhaustive search over specified parameter values using cross-validation. This ensures that we identify the best model configuration for our dataset.

Below is the implementation:


````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define parameter grid
param_grid = {
    'kernel': ['rbf', 'linear'],
    'C': [0.1, 1, 10],
    'epsilon': [0.1, 0.2, 0.3]
}

# Set up GridSearchCV
svr = SVR()
grid_search_svr = GridSearchCV(svr, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search_svr.fit(X_train, y_train)

# Best estimator and parameters
best_svr_model = grid_search_svr.best_estimator_
print("üîç Best Parameters for SVR:", grid_search_svr.best_params_)

# Predict using the best model
y_pred_svr_tuned = best_svr_model.predict(X_test)

# Evaluate performance
mse_svr_tuned = mean_squared_error(y_test, y_pred_svr_tuned)
mae_svr_tuned = mean_absolute_error(y_test, y_pred_svr_tuned)
r2_svr_tuned = r2_score(y_test, y_pred_svr_tuned)

print(f"üìä Tuned SVR Performance:\nMSE: {mse_svr_tuned:.2f}\nMAE: {mae_svr_tuned:.2f}\nR¬≤ Score: {r2_svr_tuned:.4f}")
````

##### **üîß Which hyperparameter optimization technique have you used and why?**


For **Model 3 (Support Vector Regressor - SVR)**, we used **GridSearchCV** for hyperparameter optimization.

**Why GridSearchCV?**  
- **Exhaustive Search**: GridSearchCV performs a comprehensive search over a predefined hyperparameter grid, ensuring we don‚Äôt miss potentially optimal parameter combinations.
- **Cross-Validation Integration**: It evaluates each parameter combination using k-fold cross-validation (in our case, 5-fold), providing robust performance estimates.
- **Best for Smaller Grids**: Since our parameter space (C, kernel, epsilon) was reasonably small, GridSearchCV was ideal in terms of efficiency and thoroughness.

> This approach helped us identify the best combination of SVR parameters that yielded the highest R¬≤ score while keeping the Mean Squared and Mean Absolute Errors low.
